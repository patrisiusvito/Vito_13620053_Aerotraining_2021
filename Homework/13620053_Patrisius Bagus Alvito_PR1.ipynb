```python
for i in range (1,101):
    if i%15 == 0:
        print("FizzBuzz")
    elif i%3 == 0:
        print("Fizz")
    elif i%5 == 0:
        print("Buzz")
    else:
        print(i)
```

```python
word = input("Masukan kata:")
i = len(word)-1
check = 0
for char in word:
    if char != word[i]:
        check += 1
    i-=1

if check != 0:
    print("No")
else:
    print("Yes")
```
