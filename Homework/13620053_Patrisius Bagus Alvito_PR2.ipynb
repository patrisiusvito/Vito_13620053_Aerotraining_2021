```python
class Isa:
    def __init__(self):
        self.base_alt_list = [0, 11000, 20000, 32000, 47000, 51000, 71000, 84852]
        self.lambda_list = [-0.0065, 0, 0.001, 0.0028,0, -0.0028, -0.002 ]
        self.ti_list =  [288.15,216.65,216.65, 228.65, 270.65, 270.65,214.65]
        self.atm_layer= {"Troposphere":11000, "Stratosphere":50000 , "Mesosphere": 80000,"Thermosphere":500000}

    def calc_segment(self, alt):
        self.i = 0
        for height in self.base_alt_list:
            if alt >= height and alt<= self.base_alt_list[ self.i + 1]:
                break
            else:
                self.i += 1
        return self.i

    def calc_temp(self, alt):
        segment = self.calc_segment(alt)
        base_alt = self.base_alt_list[segment]
        base_lambda = self.lambda_list[segment]
        ti = self.ti_list[segment]

        temp = ti + base_lambda*(alt-base_alt)
        return temp

    def calc_atm_layer(self, alt):
        for key in self.atm_layer:
            if alt <= self.atm_layer[key]:
                self.layer = key
                break
        return self.layer
    
```

```python

calculator = Isa()

print("Kalkulator Temperatur dan Atmosferik Layer ISA\n"
          "Valid untuk ketinggian 0<= h< 84000")

alt = int(input(" Masukkan ketinggian (meter) : "))

temp_alt = calculator.calc_temp(alt)
layer = calculator.calc_atm_layer(alt)

print(f"Tempartuer pada ketinggian {alt} m: {temp_alt} K")
print(f"Lapisan atmosfer pada ketinggian {alt} m: {layer}")
```
